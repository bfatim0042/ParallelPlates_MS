```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os
import glob

# ── parameters ────────────────────────────────────────────────────────────────
H_list   = [1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]
eta_list = [0.30, 0.35, 0.40, 0.45, 0.50, 0.55]

# directory where your .dat files live (change if needed)
DATA_DIR = "."

# colour map: one colour per eta value
cmap   = plt.cm.plasma
colors = [cmap(i / (len(eta_list) - 1)) for i in range(len(eta_list))]
# ──────────────────────────────────────────────────────────────────────────────


def load_dat(filepath):
    """Load a two-column whitespace-separated .dat file, skipping comment lines."""
    try:
        data = np.loadtxt(filepath, comments="#")
        return data[:, 0], data[:, 1]
    except Exception:
        return None, None


def plot_H(H):
    """
    For a given H, produce a figure with two columns:
      left  - in-plane g(r) for every eta
      right - density profile rho(z) for every eta
    """
    fig = plt.figure(figsize=(12, 5))
    fig.suptitle(f"$H/\sigma$ = {H:.1f}", fontsize=14, fontweight="bold")

    gs = gridspec.GridSpec(1, 2, wspace=0.35)
    ax_gr   = fig.add_subplot(gs[0])
    ax_rhoz = fig.add_subplot(gs[1])

    any_gr   = False
    any_rhoz = False

    for eta, color in zip(eta_list, colors):
        label = f"$\eta$ = {eta:.2f}"

        # ── in-plane g(r) ──────────────────────────────────────────────────
        gr_file = os.path.join(DATA_DIR, f"data/gr/gr_H{H:.1f}_eta{eta:.2f}.dat")
        # also try two-decimal H format in case filenames differ
        if not os.path.exists(gr_file):
            gr_file = os.path.join(DATA_DIR, f"data/gr/gr_H{H:.2f}_eta{eta:.2f}.dat")

        r, gr = load_dat(gr_file)
        if r is not None:
            ax_gr.plot(r, gr, color=color, linewidth=1.5, label=label)
            any_gr = True
        else:
            print(f"  [missing] {gr_file}")

        # ── density profile rho(z) ─────────────────────────────────────────
        rhoz_file = os.path.join(DATA_DIR, f"data/rhoz/rhoz_H{H:.1f}_eta{eta:.2f}.dat")
        if not os.path.exists(rhoz_file):
            rhoz_file = os.path.join(DATA_DIR, f"data/rhoz/rhoz_H{H:.2f}_eta{eta:.2f}.dat")

        z, rhoz = load_dat(rhoz_file)
        if z is not None:
            ax_rhoz.plot(z, rhoz, color=color, linewidth=1.5, label=label)
            any_rhoz = True
        else:
            print(f"  [missing] {rhoz_file}")

    # ── dress the axes ─────────────────────────────────────────────────────
    ax_gr.set_xlabel(r"$r \, / \, \sigma$", fontsize=12)
    ax_gr.set_ylabel(r"$g(r)$", fontsize=12)
    ax_gr.set_title("In-plane g(r)", fontsize=11)
    ax_gr.axhline(1.0, color="grey", linewidth=0.8, linestyle="--", alpha=0.6)
    ax_gr.set_xlim(left=0)
    ax_gr.set_ylim(bottom=0)
    if any_gr:
        ax_gr.legend(fontsize=8, framealpha=0.7)

    ax_rhoz.set_xlabel(r"$z \, / \, \sigma$", fontsize=12)
    ax_rhoz.set_ylabel(r"$\rho(z)$", fontsize=12)
    ax_rhoz.set_title(r"Density profile $\rho(z)$", fontsize=11)
    ax_rhoz.set_xlim(0, H)
    ax_rhoz.set_ylim(bottom=0)
    # mark wall positions
    ax_rhoz.axvline(0.5, color="k", linewidth=1.0, linestyle=":", alpha=0.5, label="walls")
    ax_rhoz.axvline(H - 0.5, color="k", linewidth=1.0, linestyle=":", alpha=0.5)
    if any_rhoz:
        ax_rhoz.legend(fontsize=8, framealpha=0.7)

    fname = f"plots/plot_H{H:.1f}.png"
    fig.savefig(fname, dpi=150, bbox_inches="tight")
    plt.close(fig)
    print(f"Saved {fname}")


# ── main ──────────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    for H in H_list:
        print(f"\nPlotting H = {H:.1f} ...")
        plot_H(H)

    print("\nAll done.  Files saved as plot_H*.png")
```
